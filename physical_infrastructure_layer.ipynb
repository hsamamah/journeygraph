```python
folder_path = 'data/gtfs'
pathway_df = pd.read_csv(f'{folder_path}/pathways.txt')
pathway_df

pathway_modes = {
    1 : "Walkway",
    2 : "Stairs",
    3 : "Moving sidewalk/travelator",
    4 : "Escalator",
    5 : "Elevator",
    6 : "Fare gate/turnstile"
}

pathway_df['mode_description'] = pathway_df['pathway_mode'].map(pathway_modes)

folder_path = 'data/gtfs'
stops_df = pd.read_csv(f'{folder_path}/stops.txt')

bus_stops = stops_df[pd.to_numeric(stops_df['stop_id'], errors='coerce').notnull()]
bus_stops = bus_stops.dropna(axis=1, how='all')
bus_stops

rail_stops = stops_df[pd.to_numeric(stops_df['stop_id'], errors='coerce').isnull()]
rail_stops = rail_stops.dropna(axis=1, how='all')
rail_stops
```

```python
def get_level_description(level_id):

    if pd.isna(level_id) or level_id == '':
        return 'Level Not specified'
    level_indicator = level_id.split('_')[-1]
    level_mapping = {
        'L0': 'Street Level',
        'L1': 'Mezzanine/Fare Control',
        'L2': 'Platform Level',
        'L3': 'Lower Platform/Deep Level',
        'UL1': 'Upper Level (Above Ground)',
    }
    
    return level_mapping.get(level_indicator, f'Unknown Level ({level_indicator})')

rail_stops['level_description'] = rail_stops['level_id'].apply(get_level_description)
rail_stops
```

```python
import pandas as pd
import networkx as nx

def build_graph(paths_df: pd.DataFrame, stops_df: pd.DataFrame) -> nx.DiGraph:
    """Directed graph for routing/traversal; includes all edges; respects is_bidirectional."""
    G = nx.DiGraph()
    for _, r in stops_df.iterrows():
        G.add_node(r["stop_id"], location_type=r.get("location_type"))

    for _, r in paths_df.iterrows():
        u, v = r["from_stop_id"], r["to_stop_id"]
        mode = int(r["pathway_mode"])
        bi   = int(r["is_bidirectional"])
        G.add_edge(u, v, pathway_id=r["pathway_id"], pathway_mode=mode, is_bidirectional=bi)
        if bi == 1:
            G.add_edge(v, u, pathway_id=r["pathway_id"], pathway_mode=mode, is_bidirectional=bi)
    return G

def seed_sets(stops_df: pd.DataFrame):
    """Entrances: location_type=2 and stop_id starts with ENT_. Platforms/boarding areas: 0 or 4 or empty."""
    entrances = set(
        stops_df.loc[
            (stops_df["location_type"] == 2) &
            (stops_df["stop_id"].str.upper().str.startswith("ENT_")),
            "stop_id"
        ]
    )
    platforms = set(
        stops_df.loc[
            (stops_df["location_type"].isin([0, 4])) | (stops_df["location_type"].isna()) | (stops_df["stop_id"].str.upper().str.startswith("PF_")),
            "stop_id"
        ]
    )
    return entrances, platforms

def reachable_without_modes(G: nx.DiGraph, sources: set, forbid_modes={6, 7}):
    """Forward reachability from sources, not traversing edges whose pathway_mode is 6 or 7 (fare/exit gates)."""
    sources = [s for s in sources if s in G]  
    seen = set()
    stack = list(sources)
    while stack:
        u = stack.pop()
        if u in seen:
            continue
        seen.add(u)
        for v in G.successors(u):
            if G[u][v].get("pathway_mode") in forbid_modes:
                continue
            stack.append(v)
    return seen

def build_and_partition(stops_df: pd.DataFrame, pathways_df: pd.DataFrame):

    G = build_graph(pathways_df, stops_df)

    entrances, platforms = seed_sets(stops_df)
    gate_modes = {6, 7}
    unpaid_nodes = reachable_without_modes(G, entrances, forbid_modes=gate_modes)
    paid_nodes   = reachable_without_modes(G, platforms, forbid_modes=gate_modes)

    node_side = {}
    for n in G.nodes:
        in_unpaid = n in unpaid_nodes
        in_paid   = n in paid_nodes
        if in_unpaid and not in_paid:
            node_side[n] = "UNPAID"
        elif in_paid and not in_unpaid:
            node_side[n] = "PAID"
        elif in_paid and in_unpaid:
            node_side[n] = "AMBIG"    
        else:
            node_side[n] = "UNKNOWN"  

    nodes_with_side = (
        pd.DataFrame({"stop_id": list(node_side.keys()), "side": list(node_side.values())})
        .merge(stops_df[["stop_id","location_type"]], on="stop_id", how="left")
    )

    paths = pathways_df.copy()
    paths["from_side"] = paths["from_stop_id"].map(node_side)
    paths["to_side"]   = paths["to_stop_id"].map(node_side)

    def categorize(row):
        mode = int(row["pathway_mode"])
        fs, ts = row["from_side"], row["to_side"]
        if mode == 7 and fs == "PAID" and ts == "UNPAID":
            return "exit_gate"
        if mode == 6 and fs != ts:
            return "cross_faregate"
        if fs == "UNPAID" and ts == "UNPAID":
            return "pregate_internal"
        if fs == "PAID"   and ts == "PAID":
            return "postgate_internal"
        return "mixed_non_gate"

    paths["side_category"] = paths.apply(categorize, axis=1)

    # # Buckets
    fare_boundary_edges = paths[paths["pathway_mode"].isin(gate_modes)].copy()
    pre_gate_edges      = paths[paths["side_category"] == "pregate_internal"].copy()
    post_gate_edges     = paths[paths["side_category"] == "postgate_internal"].copy()
    pre_gate_nodes      = nodes_with_side[nodes_with_side["side"] == "UNPAID"].copy()
    post_gate_nodes     = nodes_with_side[nodes_with_side["side"] == "PAID"].copy()

    # # Sanity checks
    # exit_bad_bidir = fare_boundary_edges[(fare_boundary_edges["pathway_mode"] == 7) &
    #                                      (fare_boundary_edges["is_bidirectional"] == 1)]
    # if len(exit_bad_bidir):
    #     print(f"Warning: {len(exit_bad_bidir)} exit gates are bidirectional; GTFS requires mode 7 to be unidirectional.")

    return {
        "G": G,
        "nodes_with_side": nodes_with_side,
        "paths_with_side": paths,
        "fare_boundary_edges": fare_boundary_edges,  
        "pre_gate_edges": pre_gate_edges,
        "post_gate_edges": post_gate_edges,
        "pre_gate_nodes": pre_gate_nodes,
        "post_gate_nodes": post_gate_nodes
    }


parts = build_and_partition(stops_df, pathway_df)
G = parts["G"]

parts["post_gate_nodes"]
parts["pre_gate_edges"]


full_pathway_df = parts["paths_with_side"]
```

```python
# Integrate stops and pathways data
# Merge stops info for from_stop_id
from_stops = rail_stops.rename(columns={
    'stop_id': 'from_stop_id',
    'stop_desc': 'from_stop_desc',
    'level_id': 'from_level_id',
    'location_type': 'from_location_type'
}).copy()

# Merge stops info for to_stop_id
to_stops = rail_stops.rename(columns={
    'stop_id': 'to_stop_id',
    'stop_desc': 'to_stop_desc',
    'level_id': 'to_level_id',
    'location_type': 'to_location_type'
}).copy()

# Merge pathways with from_stops and to_stops
integrated = full_pathway_df.merge(from_stops, on='from_stop_id', how='left')
integrated = integrated.merge(to_stops, on='to_stop_id', how='left')

# Display key columns for analysis
integrated[['from_stop_id', 'from_stop_desc', 'to_stop_id', 'to_stop_desc', 'pathway_id', 'mode_description', 'side_category', 'from_level_id', 'to_level_id', 'from_location_type', 'to_location_type']]
```
